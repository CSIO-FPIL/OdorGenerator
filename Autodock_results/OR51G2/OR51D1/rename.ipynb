```python
import os

# Renaming rules directly in the script
rename_map = {
    "2E_4E_Hexa_24_dienal": "trans-2-Nonenal",
    "2E_4Z_Hexa_2_4_dienal": "Z-Non-2-enal",
    "2E_Penta_2_enal": "2-Octenal",
    "246_Heptatrienal": "36-nonadienal",
    "Propionic_acid": "propionic_acid",
    "Butyric_acid": "butyric_acid",
    "Isobutyric_acid": "ISOBUTYRIC_ACID",
    "2_Methylbutanoic_acid": "2-Methylbutanoic_acid",
    "Butyl_butyrate": "ETHYL_HEXANOATE",
    "Pentyl_acetate": "ETHYL_VALERATE",
    "Hexyl_acetate": "Pentyl_propionate",
    "Heptyl_acetate": "Hexyl_acetate",
    "2_Butanol": "2-METHYL-1-BUTANOL",
    "2_Methyl_2_propanol": "Isoamyl_alcohol",
    "Isobutyraldehyde": "2-Methylbutyraldehyde",
    "Crotyl_alcohol": "cis-3-Hexen-1-ol",
    "34_Dithiahexane": "Dipropyl_disulfide",
    "13_Dithiane": "Bis-methylthio-methane",
    "245_Trithiapentane": "Dimethyl_trisulfide",
    "23_Dithiapentane": "Dimethyl_Disulfide"
}

def rename_files(directory):
    renamed_files = []
    unmatched_files = []
    missing_files = list(rename_map.keys())  # Start with all files from rename_map

    for item in os.listdir(directory):
        matched = False
        for old_name, new_name in rename_map.items():
            if old_name in item:
                prefix = item[:item.index(old_name)]
                suffix = item[item.index(old_name) + len(old_name):]
                new_item = f"{prefix}{new_name}{suffix}"
                old_path = os.path.join(directory, item)
                new_path = os.path.join(directory, new_item)
                os.rename(old_path, new_path)
                print(f"Renamed: {item} -> {new_item}")
                renamed_files.append(item)
                matched = True
                if old_name in missing_files:
                    missing_files.remove(old_name)
                break  # Stop checking other patterns once a match is found and renamed
        
        if not matched:
            unmatched_files.append(item)

    print("\nSummary:")
    print(f"Total files processed: {len(renamed_files) + len(unmatched_files)}")
    print(f"Files renamed: {len(renamed_files)}")
    print(f"Files not renamed: {len(unmatched_files)}")
    
    if unmatched_files:
        print("\nFiles that were not renamed:")
        for file in unmatched_files:
            print(f"- {file}")

    if missing_files:
        print("\nFiles from rename_map that were not found in the directory:")
        for file in missing_files:
            print(f"- {file}")

if __name__ == "__main__":
    directory = "."  # Current directory, change this if needed
    rename_files(directory)
```
